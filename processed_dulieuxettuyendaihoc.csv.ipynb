câu 4 :Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu thì điền giá trị 0

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# =========================
# 1. Khảo sát dữ liệu thiếu
# =========================

print("Tần số các giá trị trong cột DT:")
print(df['DT'].value_counts(dropna=False))

print("\nTần suất các giá trị trong cột DT:")
print(df['DT'].value_counts(normalize=True, dropna=False))

# Đếm số dữ liệu thiếu
print("\nSố lượng dữ liệu thiếu:")
print(df['DT'].isnull().sum())

# Các giá trị riêng biệt
print("\nCác giá trị riêng biệt trong cột DT:")
print(df['DT'].unique())

# =========================
# 2. Thay thế dữ liệu thiếu
# =========================

# Điền giá trị 0 vào dữ liệu thiếu
df['DT'] = df['DT'].fillna(0)

# Kiểm tra lại
print("\nSau khi thay thế dữ liệu thiếu:")
print(df['DT'].isnull().sum())

# Hiển thị 10 dòng đầu
print("\n10 dòng đầu sau khi xử lý:")
print(df[['DT']].head(10))
```

câu 5 : Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.



```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# =====================================
# 1. Thống kê dữ liệu thiếu của biến T1
# =====================================

# Bảng tần số
print("Bảng tần số của T1:")
print(df['T1'].value_counts(dropna=False))

# Bảng tần suất
print("\nBảng tần suất của T1:")
print(df['T1'].value_counts(normalize=True, dropna=False))

# Số lượng dữ liệu thiếu
print("\nSố lượng dữ liệu thiếu:")
print(df['T1'].isnull().sum())

# =====================================
# 2. Thay thế dữ liệu thiếu bằng Mean
# =====================================

# Tính giá trị trung bình
mean_T1 = df['T1'].mean()

print("\nGiá trị trung bình của T1:")
print(mean_T1)

# Thay thế dữ liệu thiếu bằng Mean
df['T1'] = df['T1'].fillna(mean_T1)

# =====================================
# 3. Kiểm tra sau khi xử lý
# =====================================

print("\nSố lượng dữ liệu thiếu sau khi thay thế:")
print(df['T1'].isnull().sum())

# Hiển thị 10 dòng đầu
print("\n10 dòng đầu sau khi xử lý:")
print(df[['T1']].head(10))
```

câu 6 : Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ==========================================
# Danh sách các biến điểm số cần xử lý
# ==========================================

cot_diem = [
    'T1','L1','H1','S1','V1','X1','D1','N1',
    'T2','L2','H2','S2','V2','X2','D2','N2',
    'T6','L6','H6','S6','V6','X6','D6','N6',
    'DH1','DH2','DH3'
]

# ==========================================
# Xử lý dữ liệu thiếu cho từng cột
# ==========================================

for cot in cot_diem:

    print(f"\n===== Cột {cot} =====")

    # Số lượng dữ liệu thiếu trước xử lý
    print("Số lượng dữ liệu thiếu trước:")
    print(df[cot].isnull().sum())

    # Tính giá trị trung bình
    mean_value = df[cot].mean()

    print("Giá trị trung bình:")
    print(mean_value)

    # Thay thế dữ liệu thiếu bằng Mean
    df[cot] = df[cot].fillna(mean_value)

    # Kiểm tra lại
    print("Số lượng dữ liệu thiếu sau:")
    print(df[cot].isnull().sum())

# ==========================================
# Hiển thị dữ liệu sau xử lý
# ==========================================

print("\n10 dòng đầu tiên sau khi xử lý:")
print(df.head(10))
```

câu 7 : Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và
12.  
• Công thức tính: TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10


```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ==========================================
# Tạo biến TBM1 (Trung bình môn lớp 10)
# ==========================================

df['TBM1'] = (
    df['T1'] * 2 +
    df['L1'] +
    df['H1'] +
    df['S1'] +
    df['V1'] * 2 +
    df['X1'] +
    df['D1'] +
    df['N1']
) / 10

# ==========================================
# Tạo biến TBM2 (Trung bình môn lớp 11)
# ==========================================

df['TBM2'] = (
    df['T2'] * 2 +
    df['L2'] +
    df['H2'] +
    df['S2'] +
    df['V2'] * 2 +
    df['X2'] +
    df['D2'] +
    df['N2']
) / 10

# ==========================================
# Tạo biến TBM3 (Trung bình môn lớp 12)
# ==========================================

df['TBM3'] = (
    df['T6'] * 2 +
    df['L6'] +
    df['H6'] +
    df['S6'] +
    df['V6'] * 2 +
    df['X6'] +
    df['D6'] +
    df['N6']
) / 10

# ==========================================
# Hiển thị kết quả
# ==========================================

print("10 dòng đầu tiên của TBM1, TBM2, TBM3:")
print(df[['TBM1', 'TBM2', 'TBM3']].head(10))
```

câu 8 : Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12 như sau:
•	Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
•	Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
•	Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
•	Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
•	Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)


```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ==========================================
# Tạo TBM1, TBM2, TBM3
# ==========================================

df['TBM1'] = (
    df['T1']*2 + df['L1'] + df['H1'] + df['S1'] +
    df['V1']*2 + df['X1'] + df['D1'] + df['N1']
) / 10

df['TBM2'] = (
    df['T2']*2 + df['L2'] + df['H2'] + df['S2'] +
    df['V2']*2 + df['X2'] + df['D2'] + df['N2']
) / 10

df['TBM3'] = (
    df['T6']*2 + df['L6'] + df['H6'] + df['S6'] +
    df['V6']*2 + df['X6'] + df['D6'] + df['N6']
) / 10

# ==========================================
# Hàm xếp loại học lực
# ==========================================

def xep_loai(tbm):

    if tbm < 5:
        return 'Y'

    elif tbm < 6.5:
        return 'TB'

    elif tbm < 8:
        return 'K'

    elif tbm < 9:
        return 'G'

    else:
        return 'XS'

# ==========================================
# Tạo các biến XL1, XL2, XL3
# ==========================================

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)

# ==========================================
# Hiển thị kết quả
# ==========================================

print(df[['TBM1','XL1','TBM2','XL2','TBM3','XL3']].head(10))
```

câu 9 : Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ==========================================
# Tạo TBM1, TBM2, TBM3
# ==========================================

df['TBM1'] = (
    df['T1']*2 + df['L1'] + df['H1'] + df['S1'] +
    df['V1']*2 + df['X1'] + df['D1'] + df['N1']
) / 10

df['TBM2'] = (
    df['T2']*2 + df['L2'] + df['H2'] + df['S2'] +
    df['V2']*2 + df['X2'] + df['D2'] + df['N2']
) / 10

df['TBM3'] = (
    df['T6']*2 + df['L6'] + df['H6'] + df['S6'] +
    df['V6']*2 + df['X6'] + df['D6'] + df['N6']
) / 10

# ==========================================
# Min-Max Normalization
# Chuyển từ thang điểm 10 sang thang điểm 4
# ==========================================

# US_TBM1
df['US_TBM1'] = (
    (df['TBM1'] - df['TBM1'].min()) /
    (df['TBM1'].max() - df['TBM1'].min())
) * 4

# US_TBM2
df['US_TBM2'] = (
    (df['TBM2'] - df['TBM2'].min()) /
    (df['TBM2'].max() - df['TBM2'].min())
) * 4

# US_TBM3
df['US_TBM3'] = (
    (df['TBM3'] - df['TBM3'].min()) /
    (df['TBM3'].max() - df['TBM3'].min())
) * 4

# ==========================================
# Hiển thị kết quả
# ==========================================

print(df[['TBM1','US_TBM1',
          'TBM2','US_TBM2',
          'TBM3','US_TBM3']].head(10))
```

câu  10 : Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
•	Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
•	Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt
•	Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược lại là rớt


```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# ==========================================
# Tạo biến KQXT
# ==========================================

def xet_tuyen(row):

    # Khối A và A1
    if row['KT'] in ['A', 'A1']:

        diem_xt = (
            row['DH1']*2 +
            row['DH2'] +
            row['DH3']
        ) / 4

    # Khối B
    elif row['KT'] == 'B':

        diem_xt = (
            row['DH1'] +
            row['DH2']*2 +
            row['DH3']
        ) / 4

    # Các khối khác
    else:

        diem_xt = (
            row['DH1'] +
            row['DH2'] +
            row['DH3']
        ) / 3

    # Xét kết quả
    if diem_xt >= 5:
        return 1
    else:
        return 0

# Áp dụng hàm cho từng dòng
df['KQXT'] = df.apply(xet_tuyen, axis=1)

# ==========================================
# Hiển thị kết quả
# ==========================================

print(df[['KT', 'DH1', 'DH2', 'DH3', 'KQXT']].head(10))
```
