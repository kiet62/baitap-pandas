Vấn đề 1

```
import pandas as pd

# Thêm header cho file CSV
column_names = [
    "Id",
    "Name",
    "Age",
    "Weight",
    "m0006",
    "m0612",
    "m1218",
    "f0006",
    "f0612",
    "f1218"
]

# Đọc dữ liệu
df = pd.read_csv(
    "patient_heart_rate.csv",
    names=column_names,
    header=None,
    on_bad_lines='skip'
)

# Hiển thị 10 dòng đầu tiên
print(df.head(10))

# Hiển thị số dòng và số cột
print("\nKích thước dữ liệu:", df.shape)
```

Vấn Đề 2

```
# Problem 2
# Tách cột Name thành Firstname và Lastname

df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)

# Xóa cột Name cũ
df = df.drop('Name', axis=1)

# Hiển thị kết quả
print(df)
```

Vấn Đề 3

```
# Problem 3
# Chuẩn hóa cột Weight về đơn vị kg

weight = df['Weight']

for i in range(0, len(weight)):
    x = str(weight[i])

    # Nếu đơn vị là lbs
    if "lbs" in x[-3:]:
        # Bỏ chữ lbs
        x = x[:-3]

        # Chuyển sang số thực
        float_x = float(x)

        # Đổi từ lbs sang kg
        y = int(float_x / 2.2)

        # Thêm đơn vị kgs
        y = str(y) + "kgs"

        weight[i] = y

print(df)
```

Vấn đề 4

```
# Problem 4
# Xóa các dòng có toàn bộ giá trị là NaN

df.dropna(how="all", inplace=True)

print(df)
```

Vấn Đề 5

```
# Problem 5
# Xóa các dòng bị trùng lặp hoàn toàn

df = df.drop_duplicates(
    subset=['Firstname', 'Lastname', 'Age', 'Weight']
)

print(df)
```

Vấn đề 6

```
# Problem 6
# Xử lý dữ liệu chứa ký tự non-ASCII

df.Firstname.replace({r'[^\x00-\x7F]+':''},
                     regex=True,
                     inplace=True)

df.Lastname.replace({r'[^\x00-\x7F]+':''},
                    regex=True,
                    inplace=True)

print(df)
```

Vấn đề 7

```
# Chuyển Weight từ "70kgs" thành 70

df['Weight'] = df['Weight'].str.replace('kgs', '', regex=False)
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

# Thống kê dữ liệu thiếu
print("Số giá trị thiếu trong Age:")
print(df['Age'].isnull().sum())

print("Số giá trị thiếu trong Weight:")
print(df['Weight'].isnull().sum())

# Xóa dòng có cả Age và Weight đều thiếu
df = df.dropna(subset=['Age', 'Weight'], how='all')

# Điền Age thiếu bằng trung bình
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

# Điền Weight thiếu bằng trung bình
mean_weight = df['Weight'].mean()
df['Weight'] = df['Weight'].fillna(mean_weight)

print(df)
```

Vấn đề 8

```
# Problem 8
# Tách các cột m0006, m0612, m1218, f0006, f0612, f1218
# thành 3 cột: PulseRate, Sex, Time

df = pd.melt(
    df,
    id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'],
    value_name='PulseRate',
    var_name='sex_and_time'
).sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

# Tách sex và time
tmp_df = df['sex_and_time'].str.extract(r'(\D)(\d+)(\d{2})', expand=True)

# Đặt tên cột tạm
tmp_df.columns = ['Sex', 'hours_lower', 'hours_upper']

# Tạo cột Time dạng 00-06, 06-12, 12-18
tmp_df['Time'] = tmp_df['hours_lower'] + '-' + tmp_df['hours_upper']

# Ghép dữ liệu
df = pd.concat([df, tmp_df], axis=1)

# Xóa các cột không cần thiết
df = df.drop(['sex_and_time', 'hours_lower', 'hours_upper'], axis=1)

# Sắp xếp lại cột
df = df[['Id', 'Age', 'Weight', 'Firstname', 'Lastname', 'PulseRate', 'Sex', 'Time']]

# Lưu file mới
df.to_csv("outputcleanup.csv", index=False)

print(df)
```

```
# Problem 11
# Xử lý dữ liệu thiếu trên biến huyết áp PulseRate

import numpy as np
import pandas as pd

# Chuyển PulseRate sang dạng số, dấu "-" sẽ thành NaN
df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')

# Khảo sát tỉ lệ dữ liệu thiếu
missing_count = df['PulseRate'].isnull().sum()
total_count = len(df)
missing_rate = missing_count / total_count * 100

print("Số giá trị PulseRate bị thiếu:", missing_count)
print("Tỉ lệ thiếu PulseRate:", missing_rate, "%")

# Hàm xử lý giá trị thiếu
def fill_pulse_rate(group):
    group = group.sort_values('Time').copy()

    for i in group.index:
        if pd.isna(group.loc[i, 'PulseRate']):

            pos = group.index.get_loc(i)

            # 1. Trung bình liền trước và liền sau
            if pos > 0 and pos < len(group) - 1:
                prev_val = group.iloc[pos - 1]['PulseRate']
                next_val = group.iloc[pos + 1]['PulseRate']

                if pd.notna(prev_val) and pd.notna(next_val):
                    group.loc[i, 'PulseRate'] = (prev_val + next_val) / 2
                    continue

            # 2. Trung bình 2 giá trị liền trước
            if pos >= 2:
                prev2 = group.iloc[pos - 2:pos]['PulseRate']

                if prev2.notna().all():
                    group.loc[i, 'PulseRate'] = prev2.mean()
                    continue

            # 3. Trung bình 2 giá trị liền sau
            if pos <= len(group) - 3:
                next2 = group.iloc[pos + 1:pos + 3]['PulseRate']

                if next2.notna().all():
                    group.loc[i, 'PulseRate'] = next2.mean()
                    continue

            # 4. Trung bình của người đó
            person_mean = group['PulseRate'].mean()

            if pd.notna(person_mean):
                group.loc[i, 'PulseRate'] = person_mean
                continue

    return group

# Áp dụng theo từng người
df = df.groupby(['Id', 'Firstname', 'Lastname'], group_keys=False).apply(fill_pulse_rate)

# 5. Nếu vẫn thiếu: thay bằng trung bình theo giới tính
df['PulseRate'] = df.groupby('Sex')['PulseRate'].transform(
    lambda x: x.fillna(x.mean())
)

# 6. Nếu vẫn thiếu: thay bằng trung bình toàn bộ dữ liệu
df['PulseRate'] = df['PulseRate'].fillna(df['PulseRate'].mean())

# Nếu cuối cùng vẫn còn thiếu thì thay bằng mức ổn định y học
df['PulseRate'] = df['PulseRate'].fillna(75)

print(df)

# Lưu file
df.to_csv("outputcleanup_final.csv", index=False)
```
